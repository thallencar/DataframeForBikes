```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# **MAIORES ÍNDICES DE ROUBOS DE BICICLETAS NO ESTADO DE SÃO PAULO**


```python
df = pd.read_csv('theftsp.csv')
```

```python
df.head()
```

```python
plt.title('Contagem de Dados por Cidade e Ano')

plt.xticks(rotation=90)
sns.countplot(x='CIDADE', hue='ANO_BO', data=df, palette='viridis', order=df['CIDADE'].value_counts().index)
plt.xlabel('Cidade')
plt.ylabel('Contagem')
```

# **ÍNDICES CLIMÁTICOS EM CORRIDAS DE BICICLETAS (GLOBAL, 2011-12)**


```python
dfr = pd.read_csv('Rentals_train.csv')
```

```python
dfr.head()
```

```python
plt.title('Climas mais favoráveis para pedalar')

plt.xticks(rotation=90)
sns.countplot(x='season', hue='temp', data=dfr, palette='viridis')
plt.xlabel('Estação do ano (1 = Inverno, 2 = Primavera , 3 = Verão , 4 = Outuno)')
plt.ylabel('Média de ciclistas')
```

# **DISTÂNCIA PERCORRIDA POR IDADE (GLOBAL, 2011-12)**


```python
dfrm = pd.read_csv('bike_buyers_clean.csv')
```

```python
dfrm.head()
```

```python
plt.title('Distância percorrida por idade')

plt.xticks(rotation=90)
sns.countplot(x='Income', hue='Commute Distance', data=dfrm, palette='viridis')
plt.xlabel('Distância')
plt.ylabel('Idade')
```
